# Introduction

**A brief overview of Regression and Classification:**

#### Classification 

Classification is the problem of machine learning in which we try to identify to which of a set of categories a new observation belongs, on the basis of a training set of data containing observations whose category membership is known. Classification is considered an instance of supervised learning, i.e., learning where a training set of correctly identified observations is available.

#### Regression

Regression is the problem of machine learning in which the machine predicts the value of the test dataset based on the training dataset. In Regression, we have one dependent variable and one or more independent variables. We try to find the "**best fitting line**" for our dataset that fits our model perfectly using Least Mean Squares Method(LMS). This is a simple case of Linear Regression, where we can predict value of any unknown data point, once we get the best fitting line. This is also considered as an instance of supervised learning.

# Review of Probability Theory

In previous section, I discussed about the Linear Algebra and why it is required in Machine Learning. Probability is equally important in Classification, Clustering and other models where we are not predicting numeric values. We decide the "goodness" of the model based on the Probability. Here, I will discuss about some key concepts which is refreshed or learnt afresh.

**Probability** is the mathematical framework for representing uncertain statements. It is the measure of the likelihood that an event will occur or not. It is measured by the ratio of the favorable cases to the whole number of cases possible. In Artificial Intelligence and Machine Learning, we are constantly dealing with uncertainty. There is very little that we can say fir sure after visualising our data. Probability theory is the Dominant framework for dealing with uncertainties. The chapter "Probability and Information Theory" was very useful and refreshed the Probability theory that I studied in my high school and undergrad. I also learnt few new concepts that we will be using in Machine Learning in this course. The chapter started with basic terminologies of Probability. A **random variable** is defined as the value of the given variable which represents the outcome of a statistical experiment. If we say that we are tossing two coins, and  we consider random variable X to be the number of Heads observed. The possible outcomes of this experiment would be {HH,HT,TH, TT}. Thus, the possible number of heads observed is no head, one head and two heads. Therefore, the possible values taken by the random variable are 0, 1, and 2 which is **discrete**. The other type of random variable is **continuous**. A continuous random variable can take infinite data points or values in the given interval. It is represented by area under the curve. A Probability Distribution is a mathematical function that provides the probabilities of occurrence of different possible outcomes in an experiment. Based on random variables Discrete and Continuous Probability Distribution are described by **Probability mass Function(PMF)** and **Probability Density Function(PDF)** respectively.  Next, I studied about the different Probability Distribution namely: **Marginal Probability** (to find subset of variables from the known set of variables), **Conditional Probability** (Probability of an event A given Probability of another event B), **Bernoulli Distribution** (for single binary random variables), **Multinoulli Distribution**, **Gaussian Distribution or Normal Distribution** (to represent random variables whose distributions are not known) , **Exponential and Laplace Distributions**. It also discussed by **Expectation** : is the mean of expected value of random variable; **Variance** gives the measure of difference between random variable x and different values of x from its probability distribution (Square root of variance is **Standard Deviation**) ;and **Covariance** : measure of how much two values are linearly related. 

**Bayes Rule** describes the probability of an event, based on prior knowledge of conditions that might be related to the event. If we know P(y|x) and need to know P(x|y) given that we also know P(x) then we can calculate P(y|x) using Bayes Rule:

$ P(x|y) = \frac{P(x) P(y|x)}{P(y)} $

Naive Bayes Classification Algorithm  are a family of simple "probabilistic classifiers" based on applying Bayes' theorem with strong (naive) independence assumptions between the features.

Sometimes, instead of using a single function to represent a probability distribution, we can split a probability distribution into many factors that we multiply together. When we represent the factorization of a probability distribution with a graph,such models are called **Structured Probablistic Model**. Models with directed edges are called **Directed Models**, while models with undirected edges are called **Undirected Models**

# Introduction for Data for Regression

This is a dataset for hourly and daily count of rental bikes between years 2011 and 2012 in Capital bikeshare system with the corresponding weather and seasonal information. Since this data was large enough and containing two individual datasets for daily and hourly count of rental bikes, I chose **hourly dataset** for my analysis. I got this data from **UCI Machine Learning repository**.

Below are the descriptive features of this dataset:

- instant: record index
- dteday : date
- season : season (1:springer, 2:summer, 3:fall, 4:winter)
- yr : year (0: 2011, 1:2012)
- mnth : month ( 1 to 12)
- hr : hour (0 to 23)
- holiday : weather day is holiday or not (extracted from [Web Link])
- weekday : day of the week
- workingday : if day is neither weekend nor holiday is 1, otherwise is 0.
+ weathersit : 
- 1: Clear, Few clouds, Partly cloudy, Partly cloudy
- 2: Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist
- 3: Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds
- 4: Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog
- temp : Normalized temperature in Celsius. The values are derived via (t-t_min)/(t_max-t_min), t_min=-8, t_max=+39 (only in hourly scale)
- atemp: Normalized feeling temperature in Celsius. The values are derived via (t-t_min)/(t_max-t_min), t_min=-16, t_max=+50 (only in hourly scale)
- hum: Normalized humidity. The values are divided to 100 (max)
- windspeed: Normalized wind speed. The values are divided to 67 (max)
- casual: count of casual users
- registered: count of registered users
- cnt: count of total rental bikes including both casual and registered

## Preprocessing Data

There seem to be missing values in the data and the fields are filled with '?'. We will need to remove those rows so that Machine Learning Algorithms can be applied properly on such datasets.

```python
#Replace '?' with NaN and then use dropna() method to delete all such rows
income_data = income_data.replace(' ?', np.NaN)
income_data = income_data.dropna()
```

```python
income_data.shape
```

Now, there is no missing or junk data in the dataset. Lets review the dataset once again.

```python
income_data
```

The income that will be predicted as part of Classification problem currently exists as <=50K or >50K. We can convert this to a binary classifier (0,1) where 0 means income <= 50K and 1 means income > 50K

```python
income_data['income'] = income_data['income'].map({' <=50K' : 0, ' >50K' : 1})
```

```python
income_data
```

Income has been successfully converted to a binary classifier (0,1), which we will use to plot and visualise the data in next step.

```python
#Separating dependent variable and independent variables for plotting
X = income_data.iloc[:,:-1]
Y = income_data.iloc[:,-1]
```

## Visualizing the data
